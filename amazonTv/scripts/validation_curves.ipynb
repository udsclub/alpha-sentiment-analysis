```python
from pandas import DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = DataFrame([['100k', 0.9418, 0.78],
                 ['200k', 0.9483, 0.81],
                 ['500k', 0.9562, 0.84],
                 ['1M', 0.9590, 0.85],
                 ['1,3M', 0.9600, 0.85]], columns = ['n_rows', 'acc', 'f1'])
b = DataFrame([['100k', 0.6874, 0.57],
                 ['200k', 0.7045, 0.60],
                 ['500k', 0.7368, 0.66],
                 ['1M', 0.7313, 0.65],
                 ['1,3M', 0.7334, 0.65]], columns = ['n_rows', 'acc', 'f1'])
```

```python
plt.figure(figsize=((16,9)))
xs = [i for i, _ in enumerate(a.acc)]
plt.plot(xs, a.acc, 'g-', label='Accuracy')
plt.plot(xs, a.f1, 'r-', label='F1 score')
# plt.axis([0, 4, 0, 1.])
plt.xlabel("number of train samples")
plt.xticks(xs, a.n_rows)
plt.title("Amazon test dataset")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,9))
xs = [i for i, _ in enumerate(b.acc)]
plt.plot(xs, b.acc, 'g-', label='Accuracy')
plt.plot(xs, b.f1, 'r-', label='F1 score')
# plt.axis([0, 4, .9, 1.])
plt.xlabel("number of train samples")
plt.xticks(xs, b.n_rows)
plt.title("Test data")
plt.legend()
plt.show()
```
